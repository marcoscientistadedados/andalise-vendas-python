```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo o arquivo
df = pd.read_csv('vendas.csv')

# Convertendo datas
df['Data'] = pd.to_datetime(df['Data'])

# Criando nova coluna de receita
df['Receita'] = df['Quantidade'] * df['Preço Unitário']

# Receita total
print('Receita Total:', df['Receita'].sum())

# Produto mais vendido
print('Mais vendido:', df.groupby('Produto')['Quantidade'].sum().idxmax())

# Receita por mês
df['Mês'] = df['Data'].dt.month
receita_por_mes = df.groupby('Mês')['Receita'].sum()

# Gráfico
receita_por_mes.plot(kind='bar', title='Receita por Mês')
plt.xlabel('Mês')
plt.ylabel('Receita')
plt.show()
```
